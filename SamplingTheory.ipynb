## Representation of a CT Signal by its Samples

ALl of the information of a continouse-time signal can be captured by a sequence of evenly spaced samples as long as:
1. The CT signal is band-limited
2. The sampling frequency is sufficiently fast

## Impulse Train Sampling
To represent a the sampling of a CT signal we can imagine the following:
$$
x_p(t) = x(t)p(t)
$$
where $x(t)$ is the CT signal being sampled 

and 

$$
p(t) = \sum_{n=-\infty}^{\infty}\delta(t-nT)
$$

Using the sampling property of the unit impulse we know that 
$$
x(t)\delta(t-t_0) = x(t_0)\delta(t-t_0)
$$

```python
#Example of unit impulse sampling
import numpy as np
import matplotlib.pyplot as plt
import my_dsp as my_dsp

N = 300
fsx = 1e3
x = my_dsp.gen_summed_sinusoid(N, fsx, np.array([20, 10, 12]), np.ones(3))

fsp = 100
Np = int(fsp * (N/fsx))
p = np.zeros(N)
p[0::Np] = 1

xp = x*p

plt.subplot(311)
plt.plot(x.real)
plt.subplot(312)
plt.stem(p)
plt.subplot(313)
plt.plot(xp, ls='none', marker='o', markevery=Np)
plt.plot(x.real, c='k', ls='--',alpha=0.25)
plt.tight_layout()
```

## Sampling Effects on Frequency Response

We can show that sampling a CT signal Creates a signal that has a periodic spectrum of infinte coppies of the origonal. 

Going Back to out model of a sampled signal:
$$
x_p(t) = x(t)p(t) = \sum_{n-\infty}^{\infty}x(nT)\delta(t-nT)
$$
Apply the Fourier Transform to $x_p(t)$ and simplifying will show that:
$$
X_p(j\omega) = \frac{1}{T} \sum_{k=-\infty}^{+\infty} X(j(\omega-k\omega_s))
$$
Which is a fancy equation that shows that the result of sampling is to create a spectrum that is an infinite sum of the origional signals spectrum where each term is shifted by $\omega_s$. Also teh whole sequece is scaled by $1/T$

```python
f_max = 2
np.set_printoptions(precision=2)
fs = 10
N = 100
freqs = np.linspace(0, f_max, f_max*fs+1)
amps = np.linspace(1,0,freqs.size)
n = np.linspace(0, N-1, N)
xp = my_dsp.gen_summed_sinusoid(N, fs, freqs, amps)
f, xpf = my_dsp.my_fft(xp, fs)
f = np.linspace(-0.5, 0.5, xpf.size)
```
