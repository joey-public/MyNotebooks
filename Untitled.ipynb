# Direct Sequence Spread Spectrum

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphabet = np.array([-1, 1], dtype=complex)
```

```python
f_symb = 1
f_chip = 10
```

```python
data = np.random.choice(alphabet, (5,1))
plt.stem(data.real)
plt.title('Data Bits')
plt.xlabel('n')
plt.ylabel('value')
```

```python
N_chips = f_chip // f_symb
chips = (2*np.random.randint(0, 2, (N_chips,))) -1
plt.stem(chips)
plt.title('Spreading Sequence')
plt.xlabel('n')
plt.ylabel('Value')
```

```python
dsss_sig=np.array([])
for bit in data:
    dsss_sig = np.hstack((dsss_sig, -1*bit*chips))
print(dsss_sig.shape)
plt.stem(dsss_sig)
```
