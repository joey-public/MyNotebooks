# Spread Spectrum

The goal of this notebook is to simulate and compare the following spread spectrum techniques.

1. Direct Sequence Spread Spectrum (DSSS)
2. Frequency Hopping Spread Spectrum (FHSS)
3. Time Hopping Spread Spectrum (THSS)
4. Chip

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Generating Data Symbols

All the spread spectrum techniques need to spread some narrowband data signal.

```python
n_symbols = 8
symbol_alphabet = np.array([-1, 1])
data_symbols = np.random.choice(symbol_alphabet, n_symbols) 
```

```python
plt.stem(data_symbols)
plt.title('TX Data Symbols')
plt.xlabel('nth Symbol')
plt.ylabel('Value')
```
