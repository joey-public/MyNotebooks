# Probability Mass Functions (PMF)

A functiom that gies the probability that a discrete random variable is exactly equal to some value. It is basically the dicrete counterpart to a Probability Density Fucntion (PDF). 

$$
p_X(x)\to P(X=x)
$$

The above realtion shows the noation ofr a PMF sequence. Basically the prbability that the random variable $X$ takes the value of $x$ is mapped into a discrete sequence $p_X(x)$

## Rules

The sum of all the elements in the PMF must equal 1, and all values of the PMF must be >= 0.



## Implementation of PMFs 

The `scipy.stats`library imcludes functions for generating many of the common PMF distribitions. 

The documentaiton can be ound here: https://docs.scipy.org/doc/scipy/reference/stats.html

Scipy creates a random variable object with helper functions.
