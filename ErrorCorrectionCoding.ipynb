# Error Correction Coding

```python
import numpy as np
import matplotlib.pyplot as plt

   
    
n_symbols = 4

bpsk_alphabet = np.array([1, -1])
qpsk_alphabet = np.array([1+1j, 1-1j, -1+1j, -1-1j])
qam16_alphabet = np.array([3+3j, 3+1j, 3-1j, 3-3j, 
                          1+3j, 1+1j, 1-1j, 1-3j, 
                          -1+3j, -1+1j, -1-1j, -1-3j,
                          -3+3j, -3+1j, -3-1j, -3-3j])

symbol_alphabet = qam16_alphabet

plt.scatter(np.real(symbol_alphabet), np.imag(symbol_alphabet))
plt.xlabel('Real')
plt.ylabel('Imag')
plt.title('Possible Data Symbols')
```

```python
a = np.random.choice(symbol_alphabet, n_symbols)
b = np.random.choice(symbol_alphabet, n_symbols)

def calc_hamming_dist(a:np.array, b:np.array)->int:
    assert a.size==b.size, 'symbol arrays must be of equal size.'
    return a.size-np.sum(np.equal(a, b))

print(f'a = {a}')
print(f'b = {b}')
print(f'Hamming Distance: {calc_hamming_dist(a,b)}')
```
